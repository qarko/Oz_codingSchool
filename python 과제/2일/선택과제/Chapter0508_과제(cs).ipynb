[문제2]

Python의 실수형(float) 계산 시 오차가 발생하는 이유에 대해 간단히 설명해 주세요


Python에서 실수형(float) 계산 시 오차가 발생하는 주된 이유는 컴퓨터의 이진 부동소수점 방식으로 숫자를 표현하기 때문입니다.  
이 방식은 유리수를 정확히 표현할 수 없기 때문에, 정밀도에 제한이 있습니다. 주요한 이유는 다음과 같습니다:  

이진 부동소수점 표현: 컴퓨터는 이진수로 소수를 표현합니다. 하지만 많은 십진 소수는 이진 소수로 정확히 표현할 수 없습니다.  
예를 들어, 0.1은 이진 소수로 정확히 표현할 수 없어서 근사값으로 저장됩니다.

유한한 비트 수: Python에서 사용하는 float 타입은 보통 64비트를 사용합니다. 이 비트 수는 표현 가능한 숫자의 범위를 결정하며,  
큰 숫자나 매우 작은 숫자의 경우 정밀도 손실이 발생할 수 있습니다.

연산 오차 누적: 소수점 이하 자릿수에서의 작은 오차가 연산을 반복하면서 누적될 수 있습니다.  
이는 계산 순서에 따라 결과가 달라질 수 있는 이유 중 하나입니다.

산술 연산 방식: 컴퓨터는 실수 연산을 정확하게 수행하기 위해 IEEE 754 표준과 같은 방식을 사용합니다.  
이 방식은 정밀도를 향상시키기 위해 다양한 기법을 사용하지만, 결국에는 유리수를 완벽하게 표현할 수 없다는 한계가 있습니다.
