[문제 2: 붕어빵 주문 받기]

손님이 주문한 붕어빵의 종류와 개수를 입력받는 기능을 구현하세요.

요구사항:
- 손님이 주문할 붕어빵의 종류와 개수를 입력받습니다.
- 주문 내용을 딕셔너리에 저장하세요.
- 주문 내역을 출력해주세요.

```python

fish = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

order = {}

while True:
    item = input("주문할 붕어빵의 종류를 입력하세요 (종료하려면 '완료' 입력): ").strip()
    if item == '완료':
        break
    if item not in fish:
        print("죄송합니다, 해당 붕어빵은 없습니다.")
        continue
    quantity = int(input(f"{item} 몇 개를 주문하시겠습니까? "))
    
    if item in order:
        order[item] += quantity
    else:
        order[item] = quantity

print("\n주문 내역:")
for item, quantity in order.items():
    print(f"{item}: {quantity}개")

# GPT 참조해서 진행 이해 안가는 부분도 조금 있음
```

[문제 3: 붕어빵 주문 처리]

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
fish = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

order = {}

while True:
    item = input("주문할 붕어빵 종류를 입력하세요 (종료하려면 '종료' 입력): ").strip()
    if item == '종료':
        break
    if item not in fish:
        print("죄송합니다, 해당 붕어빵은 없습니다.")
        continue
    quantity = int(input(f"{item} 몇 개를 주문하시겠습니까? "))
    
    if item in order:
        order[item] += quantity
    else:
        order[item] = quantity

print("\n주문 처리 결과:")
for item, quantity in order.items():
    if fish[item] >= quantity:
        fish[item] -= quantity
        print(f"{item} {quantity}개 판매 완료")
    else:
        print(f"{item} 재고 부족")

print("\n현재 재고:")
for item, stock in fish.items():
    print(f"{item}: {stock}개")
```

[문제 6: 매출 확인]

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
inventory = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

prices = {
    "팥붕어빵": 1000,
    "슈크림붕어빵": 1200,
    "초코붕어빵": 1500
}

sales = {
    "팥붕어빵": 0,
    "슈크림붕어빵": 0,
    "초코붕어빵": 0
}

def print_inventory(inventory):
    print("\n현재 재고:")
    for item, stock in inventory.items():
        print(f"{item}: {stock}개")

def print_sales(sales):
    print("\n현재 매출:")
    for item, count in sales.items():
        print(f"{item}: {count}개 판매")

def process_order(order, inventory, sales):
    total_sales = 0
    for item, quantity in order.items():
        if inventory[item] >= quantity:
            inventory[item] -= quantity
            sales[item] += quantity
            total_sales += quantity * prices[item]
            print(f"{item} {quantity}개 판매 완료")
        else:
            print(f"{item} 재고 부족")
    return total_sales

def admin_mode(inventory):
    while True:
        item = input("추가할 붕어빵 종류를 입력하세요 (종료하려면 '종료' 입력): ").strip()
        if item == '종료':
            break
        if item not in inventory:
            print("해당 붕어빵 종류는 없습니다.")
            continue
        quantity = int(input(f"추가할 {item} 재고 수량을 입력하세요: "))
        inventory[item] += quantity
        print(f"{item}의 재고가 {quantity}개 추가되었습니다.")
        print_inventory(inventory)

total_revenue = 0

while True:
    mode = input("모드를 선택하세요 (1: 주문, 2: 관리자, 3: 종료): ").strip()
    if mode == '3':
        break
    elif mode == '1':
        order = {}
        while True:
            item = input("주문할 붕어빵 종류를 입력하세요 (손님 종료: '다음 손님', 전체 종료: '종료' 입력): ").strip()
            if item == '종료':
                break
            if item == '다음 손님':
                break
            if item not in inventory:
                print("죄송합니다, 해당 붕어빵은 없습니다.")
                continue
            quantity = int(input(f"{item} 몇 개를 주문하시겠습니까? "))
            
            if item in order:
                order[item] += quantity
            else:
                order[item] = quantity
        
        if item == '종료':
            break
        
        total_sales = process_order(order, inventory, sales)
        total_revenue += total_sales
        print_inventory(inventory)
        print_sales(sales)
    elif mode == '2':
        admin_mode(inventory)

print("\n최종 매출:")
print(f"총 매출: {total_revenue}원")
print_sales(sales)
print("프로그램을 종료합니다.")

```
